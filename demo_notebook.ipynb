```python
import numpy as np

# Generate a short array of random numbers
num_points = 10
x = np.arange(num_points)  # X values (0 to num_points-1)
y = np.random.rand(num_points)  # Random Y values

# Create a scatter plot
plt.scatter(x, y, color='blue', marker='o', label='Random Points')
plt.xlabel('Index')
plt.ylabel('Random Value')
plt.title('Scatter Plot of Random Numbers')
plt.legend()
plt.show()
```
